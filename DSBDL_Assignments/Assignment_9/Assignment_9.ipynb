# Assignment 9: Data Visualization 2

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')

print(df.head())
```

```python
print(df.shape)
print(df.size)
print(df.columns)
print(df.dtypes)

df.info()
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=df)
plt.title("Age Distribution by Gender and Survival")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=df)
plt.title("Age vs Gender")
plt.show()
```

```python
sns.violinplot(x='sex', y='age', hue='survived', data=df)
plt.title("Violin Plot of Age Distribution")
plt.show()
```

```python
sns.stripplot(x='sex', y='age', hue='survived', data=df, dodge=True)
plt.title("Strip Plot")
plt.show()
```

```python
sns.swarmplot(x='sex', y='age', hue='survived', data=df)
plt.title("Swarm Plot")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=df)
sns.stripplot(x='sex', y='age', data=df, color='black', alpha=0.3)
plt.title("Boxplot + Data Points")
plt.show()
```

```python
df['survived'].value_counts().plot.pie(
    autopct='%1.1f%%',
    labels=['Not Survived', 'Survived'],
    startangle=90
)
plt.title("Survival Distribution")
plt.ylabel('')
plt.show()
```

```python
df['sex'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Gender Distribution")
plt.ylabel('')
plt.show()
```

```python
df['pclass'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Passenger Class Distribution")
plt.ylabel('')
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
